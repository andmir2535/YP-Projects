### Подготовка данных


```python
data.columns = data.columns.str.lower() #Приведём данные к нижнему регисттру.
```

```python
display(data.isna().sum()) #Посмотрим сколько пропущенных значений.
```

Замечаем, что большая часть пропусков, в столбцах **critic_Score**, **rating**, **user_score**. Заменить какими-либо значения эти данные не удастся, иначе мы столкнёмся с искажением данных. А значения: название игры, жанр и платформа, в общей совокупности занимают не более 1.7% данных. Эти данные не повлияют на ход исследования, их можно удалить.

```python
data[data['name'].isna()] #Перед удалением посмотрим на эти строки (name)
```

```python
data[data['genre'].isna()] #Перед удалением посмотрим на эти строки (genre)
```

```python
data[data['platform'].isna()] #Перед удалением посмотрим на эти строки (platform)
```

```python
data[data['year_of_release'].isna()] #Перед удалением посмотрим на эти строки (year_of_release)
```

```python
data = data.dropna(subset=['name']).reset_index(drop=True)
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
data = data.dropna(subset=['genre']).reset_index(drop=True)
data = data.dropna(subset=['platform']).reset_index(drop=True)
```

```python
data[data['user_score'].notna()]['user_score'].sort_values().unique()
```

В столбце Оценка Пользователей, есть значение 'tbd' (определить позже), мы же понимаем это как пропуски в данных. Заменим эти данные на NAN

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = float('nan')
```

Пропуски в рейтинге заменим на ND (no data) - нет данных. А аббревиатуру **K-A**, заменим на современную **E**.

```python
data['rating'] = data['rating'].fillna(value='ND')
data['rating'] = data['rating'].replace('K-A', 'E')
```

Годы выпуска заменим на целые числа. А заменённые пропуски в оценках пользователей приведём к дробным значениям. 

```python
data['year_of_release'] = data['year_of_release'].astype(int)
data['user_score'] = data['user_score'].astype(float)
```

У нас есть данные продаж, в 4 условно-разделённых регионах. Создадим общий столбец, подсчитывающий продажи по всему миру.
Назовём его **world_sales**.

```python
#data['world_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales'] 
```

```python
data['world_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
```

В дальнейшем анализе нам понадобятся, названия компаний-разработчиков домашних игровых консолей, мобильных платформ и персональных компьютеров. Для этого нам нужно упорядочить данные.

```python
data[data['platform'].notna()]['platform'].sort_values(ascending=False).unique() #Посмотрим, какие платформы у нас есть
```

Ссылаясь на личный опыт, а также на информацию из открытых источников, распределим данные платформы на 3 типа. **PC** - персональный компьютер, **mobile** - мобильные, переносные платформы, **console** - домашняя консоль.

```python
def platform_type(platform):
    mobile = ['3DS', 'DS', 'GB', 'GBA', 'GG', 'PSP', 'PSV']
    console = ['XOne', 'XB', 'X360', 'WiiU', 'Wii', 'WS', 'TG16', 'SNES', 'SCD', 'SAT', 'PS4', 'PS3', 'PS2', 'PS', 'PCFX', 'NG', 'NES', 'N64','GEN', 'GC', 'DC', '3DO', '2600']
    if platform == 'PC':
        platform_type = 'PC'
    if platform in mobile:
        platform_type = 'mobile'
    if platform in console:
        platform_type = 'console'
    return platform_type
data['platform_type'] = data['platform'].apply(platform_type)
    
```

Теперь распределим платформы по производителям.

```python
def manufacturer(platform):
    company = {'Atari' : ['2600'],
             'Bandai' : ['WS'],
             'Panasonic' : ['3DO'],
             'Microsoft' : ['XOne', 'XB', 'X360'],
             'Nintendo' : ['3DS', 'DS', 'GB', 'GBA', 'GC', 'N64', 'NES', 'SCD', 'SNES', 'Wii', 'WiiU'],
             'NEC' : ['PCFX', 'TG16'],
             'Nokia' : ['NG'],
             'Sega' : ['DC', 'GEN', 'GG', 'SAT'],
             'Sony' : ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV'],
             'PC' : ['PC']
            }
    for key, value in company.items():
        if platform in value:
            return key
data['company'] = data['platform'].apply(manufacturer)
```

Посмотрим на корреляцию актуальных платформ.

```python
for p in actual_platforms:
    print(p)
    data_select = data.query('platform == @p')
    print(f'Корреляция между оценкой критиков и продажами: {data_select.critic_score.corr(data_select.world_sales):.2}')
    print(f'Корреляция между оценкой пользователей и продажами: {data_select.user_score.corr(data_select.world_sales):.2}')
    print(f'Корреляция между оценками критиков и пользователей: {data_select.critic_score.corr(data_select.user_score):.2}')
    print('--' * 30)
```

Ни одно из значений корреляций между оценкой критиков и продажами нельзя назвать сильным. У оценок пользователей и продажах, значения минимальны, а у некоторых платформ и, вовсе, отрицательные. Единственное исключение, консоль WiiU, у которой значение 0.4, возможно, это значение отличается от всех остальных из-за специфики коммьюнити и аудитории данного гаджета. Корреляция между оценками критиков и пользователей в случаях **PSV, 3DS, WiiU** достаточно высокая - это говорит о том, что чаще всего пользователи и критики ставят одинаковые оценки. В остальных актульных платформах уровень корреляции колеблеться от 0.4 до 0.56, однозначно утвержать, о схожести оценок пользователей и критиков нельзя.

### Проверка гипотез

Обе рассматриваемые гипотезы являются двусторонними, поэтому для рассмотрения гипотез применим метод Стьдента. Выборки, которые мы рассматриваем не зависят друг от друга, поэтому для проверки гипотез воспользуемся методом **ttest_ind**. Этот метод рассматривает гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим данную гипотезу. **Нулевая гипотеза - "средние рейтинги одинаковые"**.  Значение альфа, возьмём 5%.

```python
from scipy import stats as st #Ипортируем библиотеку scipy для высчитывания гипотез.
```

```python
alpha = 0.05  # критический уровень статистической значимости
results = st.ttest_ind(
    data_actual[(data_actual.platform == "XOne") & (data_actual.user_score.notna())]['user_score'], 
    data_actual[(data_actual.platform == "PC") & (data_actual.user_score.notna())]['user_score']
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 

display(data_actual.query('platform in ["XOne", "PC"]')
        .pivot_table(index='platform', values='user_score', aggfunc=['count', 'mean', 'median', 'var']))
```

Гипотеза верна - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза - "средние рейтинги одинаковые".

```python
alpha = 0.05  # критический уровень статистической значимости
results = st.ttest_ind(
    data_actual[(data_actual.genre == "Action") & (data_actual.user_score.notna())]['user_score'], 
    data_actual[(data_actual.genre == "Sports") & (data_actual.user_score.notna())]['user_score']
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
    
display(data_actual.query('genre in ["Action", "Sports"]')
        .pivot_table(index='genre', values='user_score', aggfunc=['count', 'mean', 'median', 'var']))
```

Гипотеза опровергнута, что также доказано таблицей. Средние пользовательские рейтинги жанров Action и Sports разные.
